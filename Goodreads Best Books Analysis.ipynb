# Part 1 - Cleaning BBE List Dataset

```python
#importing required libraries
import pandas as pd
import numpy as np
```

#### Datasource: Lorena Casanova Lozano, & Sergio Costa Planells. (2020). Best Books Ever Dataset (1.0.0) [Data set]. Zenodo. https://doi.org/10.5281/zenodo.4265096


```python
#importing the dataset
df = pd.read_csv("D:/projects/Goodreads/Best_Books_Ever.csv")
```

```python
#having a peek at the dataset and seeing if it is loaded correctly 
df.head()
```

```python
print('There are ' + str(len(df.axes[0])) + ' rows and ' + str(len(df.axes[1])) + ' columns present in the dataset.')
```

```python
#checking the datatypes to see if there is any mismatch
df.info()
```

```python
#removing the title from book id column 
df['bookId'] = df.bookId.str.split('.').str[0]
df['bookId'] = df.bookId.str.split('-').str[0]
df['bookId']
```

```python
#diving the genre column into primary and seconday genre based on postions
df['primary genre'] = df.genres.str.split(',').str[0]
df['secondary genre'] = df.genres.str.split(',').str[1]

#keeping main author in the author column
df['author'] = df['author'].str.split(',').str[0]
```

```python
#counting the number of awards won by each book for further analysis on Power BI
df['awards_won'] = df['awards'].str.count(',')
```

```python
#checking for shape integrity 
df.shape
```

###### Splitting the ratingsbyStars column into RatingDist1, RatingDist2, RatingDist3, RatingDist4 and RatingDist5 that refer to the number of 1-5 star ratings that are left by the users.

```python
df[['RatingDist5', 'RatingDist4', 'RatingDist3', 'RatingDist2','RatingDist1']] = df['ratingsByStars'].str.split(',', expand = True)
```

```python
#dropping columns that wont be used for the analysis
col = ["genres","characters","setting","ratingsByStars","coverImg"]
df.drop(col,axis=1, inplace=True)
```

```python
#futher cleaning the strings 

df['primary genre']= df['primary genre'].str.strip("[''")
df['RatingDist5']= df['RatingDist5'].str.strip("[''")
df['RatingDist1']= df['RatingDist1'].str.strip("' ']")
df['secondary genre']= df['secondary genre'].str.strip("'']")

strip = ['RatingDist4', 'RatingDist3', 'RatingDist2','secondary genre']
for i in strip:
    df[i] = df[i].str.strip("' '")
```

```python
#checking for duplicates
df[df.duplicated()]
```

```python
#since there are duplicated values present in the dataset, we will need to remove these in order to make the data cleaner as 
#these duplicated values are probably the same books republished under a different publication
df.drop_duplicates(inplace=True);
```

```python
df['publishDate']
```

#### It can be seen that the date column is very unclean which needs to be treated before pushing the data to PowerBI


```python
df['publishDate'] = df['publishDate'].str.replace('January ', '01/')
df['publishDate'] = df['publishDate'].str.replace('February ', '02/')
df['publishDate'] = df['publishDate'].str.replace('March ', '03/')
df['publishDate'] = df['publishDate'].str.replace('April ', '04/')
df['publishDate'] = df['publishDate'].str.replace('May ', '05/')
df['publishDate'] = df['publishDate'].str.replace('June ', '06/')
df['publishDate'] = df['publishDate'].str.replace('July ', '07/')
df['publishDate'] = df['publishDate'].str.replace('August ', '08/')
df['publishDate'] = df['publishDate'].str.replace('September ', '09/')
df['publishDate'] = df['publishDate'].str.replace('October ', '10/')
df['publishDate'] = df['publishDate'].str.replace('November ', '11/')
df['publishDate'] = df['publishDate'].str.replace('December ', '12/')
df['publishDate'] = df['publishDate'].str.replace('th ', '/')
df['publishDate'] = df['publishDate'].str.replace('st ', '/')
df['publishDate'] = df['publishDate'].str.replace('nd ', '/')
```

#### There is no Language column in this extarct, which is why I will be extracting the Languages from ISBN code of the books and matching it with internationally accepted keys for particular prefixes. 
##### [source](https://en.wikipedia.org/wiki/List_of_ISBN_registration_groups#Prefixes_of_length_1)

```python
#As the first and first two digits of an ISBN Code of a book are mapped to a particular language, I am therefore extracting 
#those digits, ignoring the extra characters present in the column 
df_Mb['ISBN_RG'] = np.where(df_Mb.ISBN.str[2] == '8', df_Mb['ISBN'].astype(str).str[2:4], 
            np.where(df_Mb.ISBN.str[2] == '9', df_Mb['ISBN'].astype(str).str[2:4], 
            df_Mb['ISBN'].astype(str).str[2]))
```

###### keeping the data for books with ISBN RG with maximum 2 digits as mapping for those is available

```python
df_1 = pd.read_excel("D:/projects/Goodreads/ISBN RG.xlsx",sheet_name= 'Prefixes of length 1_edit_')
df_2 = pd.read_excel("D:/projects/Goodreads/ISBN RG.xlsx",sheet_name= 'Prefixes of length 2_edit_')
```

```python
#merging the two datasets
df_1 = pd.concat([df_1,df_2],axis =0)
df_1
```

```python
df_1.drop("ISBN example_x000D_\n(book)", axis=1, inplace=True)
df_1.drop("Type", axis=1, inplace=True)
```

```python
df_1.reset_index(inplace = True)
df_1.drop("index", axis=1, inplace=True)
```

```python
df_1 = df_1[~df_1['Country, region, or language area'].str.contains("see")]
df_1 = df_1[~df_1['Country, region, or language area'].str.contains("unallocated")]
```

```python
#Renaming the columns to make it easier to merge with the main dataset later
df_1.rename(columns = {'Registration group':'ISBN_RG','Country, region, or language area':'Language'}, inplace = True)
df_1.info()
```

```python
# df_1['ISBN_RG'] = df_1['ISBN_RG'].astype(str)
# df_1.info()
```

```python
df_1.reset_index(inplace = True)
df_1.drop("index", axis=1, inplace=True)
df_1;
```

```python
#merging the ISBN dataset with the main dataset
df_Mb = pd.merge(df_Mb,df_1,on = 'ISBN_RG',how = 'left')
```
